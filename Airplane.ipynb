```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
missing_values = ['N/a','NaN',np.nan]
df = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908.csv')
df
```

```python
#To Check the Null Values in Data Set
df.isnull().sum()
```

```python
#To viaualize Null Values in Data 
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
#To check if there is any Empty Row
df.dropna(how='all')
```

```python
df_copy = df.copy()
```

```python
df_copy['Time'] = df_copy['Time'].fillna('00:00')
df_copy['Location'] = df_copy['Location'].fillna('Pune')
df_copy['Operator'] = df_copy['Operator'].fillna('Air India')
df_copy['Flight #'] = df_copy['Flight #'].fillna(0000)
df_copy['Route'] = df_copy['Route'].fillna('Test Flight')
df_copy['Type'] = df_copy['Type'].fillna('Air Bus')
df_copy['Registration'] = df_copy['Registration'].fillna('P-PUN')
df_copy['cn/In'] = df_copy['cn/In'].fillna(1)
df_copy['Aboard'] = df_copy['Aboard'].fillna(10)
df_copy['Fatalities'] = df_copy['Fatalities'].fillna(1.0)
df_copy['Ground'] = df_copy['Ground'].fillna(0.0)
df_copy['Summary'] = df_copy['Summary'].fillna('Case Under Investigation')
```

```python
df_copy.isnull().sum()
```

```python
sns.heatmap(df_copy.isnull(),yticklabels=False)
```
